## Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2

plt.figure(figsize=(12,7))
plt.plot(x, y, c = 'r', ls = '-', lw = 3, marker = 'o', markersize = 12, markeredgecolor = 'b', 
         markerfacecolor = 'm', label = '$y = x^2$')

plt.legend(fontsize = 12, loc = 'upper center')
plt.grid(c = 'k', ls = '--', alpha = 0.5)

plt.title('Function $Y=X^2$\nin range (-10, 10)', fontsize = 18, fontweight = 2, fontname = 'Times New Roman',
          c = 'darkblue')
plt.xlabel('X range', fontsize = 13, fontweight = 2, fontname = 'Times New Roman')
plt.ylabel('Y range', fontsize = 13, fontweight = 2, fontname = 'Times New Roman')

#plt.text(-5, 80, 'first text\nin figure', fontsize = 13, fontweight = 2, fontname = 'Times New Roman')

#plt.xlim(left = 0, right = 9)
#plt.ylim(top = 80, bottom = 10)

plt.axhline(40, c = 'k', ls = '--')
plt.axvline(0, c = 'g', ls= '-.')

plt.savefig('text_1.jpg', dpi = 700)
plt.show()
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2
plt.figure(figsize=(12,7))
plt.plot(x, 10*y, c = 'r', ls = '-', lw = 3 ,label = '$y = x^2$')
plt.plot(x, x**3, c = 'g', ls = '-.', lw = 3 ,label = '$y = x^3$')
plt.plot(x, 150*np.sin(x), c = 'b', ls = '--', lw = 3 ,label = '$y = sin(x)$')

plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.plot(x, x**2)
plt.grid()

plt.subplot(2,3,2)
plt.plot(x, x**3)

plt.subplot(2,3,3)
plt.plot(np.sin(x))

plt.subplot(2,3,5)
plt.plot(x, np.tan(x))

plt.show()
```

$$ y = 20 + x $$
$$ y = 100 - 2x $$

```python
x = np.linspace(0, 100, 100)
y_1 = 20 + x
y_2 = 100 - 2 * x

plt.figure(figsize=(12,7))
plt.plot(x, y_1, c='r', lw = 2, label = 'Supply')
plt.plot(x, y_2, c='b', lw = 2, label = 'Demand' )

plt.plot(x, y_2-20,  c='g', lw = 2,label = 'Demand after taxation' , ls = '-.')

plt.axhline(80/3+20, c = 'k', ls = '--', lw = 1.5)
plt.axvline(80/3, c = 'k', ls = '--', lw = 1.5)

plt.xlim(left = 0, right= 100)
plt.ylim(bottom = 0)

plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Demand and Supply plot')

plt.legend(fontsize = 12)
plt.grid()

plt.fill_between(x, y_2, y_1, where = (y_2 > y_1) & (x >= 20) , color = 'darkorange', alpha = 0.7)

plt.show()
```

## Bar

```python
plt.figure(figsize=(12,7))

plt.bar([1,3,5,7], [5,8,2,6], label = 'ONE')
plt.bar([2,4,6,8], [1,4,8,3], label = 'TWO')

plt.legend(fontsize = 13, loc = 'upper left')
#plt.grid()
plt.show()
```

## Scatter

```python
x = np.linspace(0,10, 100)
e = np.random.rand(100) * 100
y = x + e

scale = np.random.rand(100)*800
color = np.random.rand(100)

plt.figure(figsize=(12,7))
plt.scatter(x,y, s = scale, c = color)

plt.colorbar().set_label('color bar', fontsize = 12)

plt.grid()
plt.show()
```

```python
z = np.arange(100)
d = np.arange(100)*2
m = np.arange(100)

a = np.random.randn(100)
b = z + a * 10

plt.figure(figsize=(12,7))
plt.scatter(z, b, c = m, s = d)

plt.colorbar()
plt.grid()
plt.show()
```

## Histogram

$$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $$

```python
data = np.random.normal(0,1,100000)

x = np.linspace(-5, 5, 100)
y = 1 / np.sqrt(2*np.pi)*np.exp(-x**2/2)

plt.figure(figsize=(12,7))
plt.hist(data, 70, density=1)
plt.plot(x, y, lw=3, c= 'k')

plt.show()
```
