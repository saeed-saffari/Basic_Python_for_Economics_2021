## Merge

```python
left = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A'  : ['A0', 'A1', 'A2', 'A3'],
    'B'  : ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k4'],
    'C'  : ['C0', 'C1', 'C2', 'C4'],
    'D'  : ['D0', 'D1', 'D2', 'D4']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how = 'inner', on = 'key')
```

```python
pd.merge(left, right, how = 'outer', on = 'key')
```

```python
pd.merge(left, right, how = 'left', on = 'key')
```

```python
pd.merge(left, right, how = 'right', on = 'key')
```

## Group By

```python
data = {
    'Company': ['GOOG', 'GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Person' : ['Sam', 'Charlie','John', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
    'Sales'  : [200, 120, 248, 340, 124, 243, 350]
}
```

```python
df = pd.DataFrame(data = data)
df
```

```python
by_comp = df.groupby('Company')
```

```python
by_comp
```

```python
by_comp.mean()
```

```python
by_comp.std()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
from pandas_datareader import wb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
inf = wb.download(indicator='FP.CPI.TOTL.ZG',country = 'all', start = 1960, end = 2020)
```

```python
inf.head()
```

```python
inf.unstack().head()
```

```python
inf2 = inf.unstack()
```

```python
inf2.columns = np.arange(1960, 2021)
inf2.head()
```

```python
inf2.index
```

```python
inf3 = inf2.T
inf3.head()
```

```python
inf3.index.rename('year', inplace=True)
```

```python
inf3.head()
```

```python
inf3['Iran, Islamic Rep.'].plot(figsize = (15,8), lw = 3)
plt.title('Iran inflation', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('INflation rate', fontsize = 12)
plt.grid()
plt.show()
```

```python
inf3[['Iran, Islamic Rep.', 'Norway', 'Canada']].plot(figsize = (15,8), lw = 3)
plt.title('Iran inflation', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('INflation rate', fontsize = 12)
plt.grid()
plt.show()
```

### Second data

```python
unem = wb.download(indicator= 'SL.UEM.TOTL.ZS', country = 'all', start = 1960, end = 2020)
```

```python
unem.head()
```

```python
unem2.head()
```

```python
unem2.columns = np.arange(1960, 2021)
```

```python
unem3 = unem2.T
```

```python
unem3[['Iran, Islamic Rep.', 'Norway', 'Canada']].plot(figsize = (15,8), lw = 3)
plt.title('Iran Unemployment', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Unemployment rate', fontsize = 12)
plt.grid()
plt.show()
```
