## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, y, c = 'r', lw = 3, marker = 'o', markersize = 15 , markeredgecolor = 'b',
         markerfacecolor = 'm' , label = '$Y = x^2$')

plt.legend(fontsize = 12, loc = 'upper center')
plt.grid(c = 'k', ls = '--', alpha = 0.5)

plt.xlabel('X range', fontsize = 15, fontweight = 2, fontname = 'Times New Roman')
plt.ylabel('Y range',  fontsize = 15, fontweight = 2, fontname = 'Times New Roman')
plt.title('Function $Y = X^2$\n',  fontsize = 20, fontweight = 2, fontname = 'Times New Roman', c = 'darkblue')

#plt.ylim(top = 80, bottom = 30)
#plt.xlim(left = 0, right = 8)

#plt.savefig('test_2.jpg', dpi = 500)

plt.show()
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, 10*y, 'r', label = '$y = x^2$', lw = 3)
plt.plot(x, x**3, 'b', label = '$y = x^3$', lw = 3)
plt.plot(x, 150* np.sin(x), 'g', label = '$y = sin(x)$', lw = 3)

plt.grid()
plt.legend(fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(2, 3, 1)
plt.plot(x, x**2)

plt.subplot(2,3,2)
plt.plot(x, x**3)

plt.subplot(2,3, 3)
plt.plot(x, np.sin(x))

plt.subplot(2,3,5)
plt.plot(x, np.tan(x))

plt.show()
```

## Bar

## Scatter

```python
x = np.linspace(0, 100, 100)
e = np.random.rand(100) * 100
y = x +  e
scale = np.random.rand(100) * 1000
color = np.random.rand(100)

plt.figure(figsize=(15,8))
plt.scatter(x, y, s = scale, c = color)

plt.colorbar().set_label('color bar', fontsize = 15)

plt.grid()
plt.show()
```

## Histogram

$$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $$

```python
data = np.random.normal(0, 1, 1000000)
```

```python
x = np.linspace(-4, 4, 100)
y = 1 / np.sqrt(2*np.pi) * np.exp(-x**2 /2)

plt.figure(figsize=(15,8))
plt.hist(data,100,  density=1)
plt.plot(x, y , lw = 3, c = 'k')

plt.show()
```
